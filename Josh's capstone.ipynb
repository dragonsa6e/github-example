```python
import requests
from bs4 import BeautifulSoup
```

Going to download the data from Wikipedia in order to start to create the dataframe for the project. I will also display the entires so I can get an idea of the data that I have downloaded. 

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
res = requests.get(url)
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
wiki_df=  pd.read_html(str(table))[0]

wiki_df.head()
```

Next I'm going to remove entries that the borough are not assigned

```python
df=wiki_df[wiki_df.Borough !='Not assigned']
df=df.reset_index(drop = True)

df.head()
```

Now going to sort the neighborhoods together by the same postcode

```python
df=df.sort_values('Postcode', axis=0)
df=df.reset_index(drop = True)
df
```

Now I'm going to group the Neighbourhoods together. 

```python
df=df.groupby(['Postcode','Borough'], as_index=False).sum()
df.head()
```

```python
df.shape
```

To now add the Latitude and Longitude to the dataframe. Also I will change the column's name to get prepared for the merging. 

```python
lat_long=pd.read_csv('http://cocl.us/Geospatial_data', header=0, names=('Postcode','Latitude','Longitude'))
lat_long
```

Now I'm going to merge my current data frame with the latitude and longitude dataframe

```python
master_df=df.join(lat_long.set_index('Postcode'),on='Postcode')
master_df
```
